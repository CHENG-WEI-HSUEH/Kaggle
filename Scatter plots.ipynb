```python
#Preprocess
import pandas as pd
#Visualization
import matplotlib.pyplot as plt
#Array package
import numpy as np

addr = "https://raw.githubusercontent.com/CHENG-WEI-HSUEH/Kaggle/main/CardioGoodFitness.csv"
raw_data = pd.read_csv(addr)


# Add title and axis names
plt.title('The relation of usage and miles by models')
plt.xlabel('Usage per week')
plt.ylabel('Miles ber week')
plt.xticks(np.arange(0,5,1))
plt.yticks(np.arange(15, 50, 5))

x = raw_data.groupby('Product').get_group('TM195')['Fitness'].astype(np.int)
y = raw_data.groupby('Product').get_group('TM195')['Age'].astype(np.int)
plt.scatter(x, y, color = 'hotpink')

#x = raw_data.groupby('Product').get_group('TM498')['Fitness'].astype(np.int)
#y = raw_data.groupby('Product').get_group('TM498')['Miles'].astype(np.int)
#plt.scatter(x, y, color = '#88c999')
'''





plt.scatter(x, y)
plt.show()'''
```
